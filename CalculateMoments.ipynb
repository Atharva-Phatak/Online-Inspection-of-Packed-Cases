```python
import numpy as np
import glob
import cv2
import os
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
train_path=r"C:/Users/Asus/Desktop/sih2/ripeness/C"
train_names=os.listdir(train_path)
#print(train_names)
```

```python
dframe= pd.DataFrame(columns = ['mean_H' ,'mean_S','mean_V','stddev_H' ,'stddev_S','stddev_V','entropy_H', 'entropy_S' ,'entropy_V','Skew_H','Skew_S','Skew_V','Kurtosis_H','Kurtosis_S','Kurtosis_V','label'] , index = [im for im in train_names])
a=[]
hlist = []
for train_name in train_names:
    cur_path = train_path+ "/" + train_name
    cur_label = train_name
    i = 1
    for file in glob.glob(cur_path ):#+ "/*.jpg"):
        
        #print("Processing Image - {} in {}".format(i, cur_label))
        img = cv2.imread(file)
        (mean , std) = cv2.meanStdDev(img)
        img = cv2.cvtColor(cv2.imread(train_path+'/'+train_name,1),cv2.COLOR_RGB2HSV)
        r,c,ch = img.shape
        mean,std = cv2.meanStdDev(img)
        ent = []

        for i in range(ch):
          norm_dist = np.random.normal(mean[i],std[i],r*c)
          h1,_ = np.histogram(norm_dist,bins = 256,range = (0,256))
          ent.append(stats.entropy(h1))
        skew = [stats.skew(i.flatten()) for i in cv2.split(img)]
        kurt = [stats.kurtosis(j.flatten()) for j in cv2.split(img)]
        dframe.loc[train_name,'mean_H']=mean[0][0]
        dframe.loc[train_name,'mean_S']=mean[1][0]
        dframe.loc[train_name,'mean_V']=mean[2][0]
        dframe.loc[train_name,'stddev_H']=std[0][0]
        dframe.loc[train_name,'stddev_S']=std[1][0]
        dframe.loc[train_name,'stddev_V']=std[2][0]
        dframe.loc[train_name,'entropy_H']=ent[0]
        dframe.loc[train_name,'entropy_S']=ent[1]
        dframe.loc[train_name,'entropy_V']=ent[2]
        dframe.loc[train_name,'Skew_H']=skew[0]
        dframe.loc[train_name,'Skew_S']=skew[1]
        dframe.loc[train_name,'Skew_V']=skew[2]
        dframe.loc[train_name,'Kurtosis_H']=kurt[0]
        dframe.loc[train_name,'Kurtosis_S']=kurt[1]
        dframe.loc[train_name,'Kurtosis_V']=kurt[2]
        
        np.array(a.append(mean))
        i = i + 1
```
